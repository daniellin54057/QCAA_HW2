```python
import itertools
import time

weights = [23,31,29,44,53,38,63,85,89,82]
values  = [92,57,49,68,60,43,67,84,87,72]
W = 165

best_value = 0
best_combo = None
best_weight = 0

# -----------------------------
# Start timing
# -----------------------------
start = time.time()

for bits in itertools.product([0,1], repeat=10):
    total_w = sum(w*b for w,b in zip(weights,bits))
    total_v = sum(v*b for v,b in zip(values,bits))
    
    if total_w <= W and total_v > best_value:
        best_value = total_v
        best_combo = bits
        best_weight = total_w

# -----------------------------
# End timing
# -----------------------------
elapsed = time.time() - start

print("best combination:", best_combo)
print("selected items:", [i+1 for i,b in enumerate(best_combo) if b == 1])
print("total weight:", best_weight)
print("total value:", best_value)
print("computation time:", elapsed, "seconds")
```
